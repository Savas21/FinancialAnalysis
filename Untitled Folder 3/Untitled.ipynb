```python
# clean the data
gld = pdr.get_data_yahoo('GLD', start='2013-01-01')
gld = gld.drop('Adj Close', axis=1) # drop adj close colm
gld =round(gld,2) # round by 2 digit decimal 
gld.head()
```

```python
# take a qucik look to data
with plt.style.context('ggplot'):
    plt.plot(gld.Close)
```

```python
# down sample the data
gld.index = pd.DatetimeIndex(gld.index)
# get last Business day of the Mont (BM) or (M)ontht
montly = gld.resample('BM').last()# get last price of the month
montly.head()
```

```python
plt.plot(montly.Close)
```

Take A look at some Basic Transformation

```python
gld['Change'] = gld['Close'] - gld['Close'].shift() # by default it looks 1 day 
```

```python
gld.head()
```

```python
# scale change to the prive by percent return using instantanious rate of return
gld['LN_change'] = np.log(gld['Close']/ gld['Close'].shift())
gld.head()
```

```python
with plt.style.context('ggplot'):
    plt.figure(figsize=(10,8))
    plt.hist(gld.LN_change[1:],bins=50,edgecolor='black', density='True')
```

```python
# Gold`s volatility 
gld['Daily Vol'] = gld['LN_change'].rolling(21).std().shift()
```

```python
gld['Exp Change'] = gld['Close'] * gld['Daily Vol']
gld.head(23)
```
